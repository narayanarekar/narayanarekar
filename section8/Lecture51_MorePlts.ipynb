```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set()
## re-install seaborn
##conda install -c conda-forge seaborn
```

```python
n = pd.read_csv("nobel.csv",parse_dates=True)
```

```python
n.head()
```

### Nobel prize across the years

```python
year=n['Year'].value_counts()
```

```python
sns.lineplot(x=year.index,y=year.values,color='blue')

plt.xticks(rotation=90)
plt.title('Nobel Prizes by Year')
plt.show()
```

### Temporal trends across categories

```python
year_cat=n.groupby(['Year','Category'])['Laureate ID'].count().reset_index()
```

```python
year_cat
```

```python
g = sns.FacetGrid(year_cat, col='Category', hue='Category', col_wrap=4, )
g = g.map(plt.plot, 'Year', 'Laureate ID')
g = g.map(plt.fill_between, 'Year', 'Laureate ID', alpha=0.2).set_titles("{col_name} Category")
g = g.set_titles("{col_name}")

 
plt.show()
```

### Horizontal barplot

```python
sns.countplot(y="Birth Country", 
              data=cstat,
              order=cstat["Birth Country"].value_counts().index,
              palette='Blues_r')
sns.set(rc={"figure.figsize": (10, 12)})
```

```python
sns.countplot(y="Region", 
              data=cstat,
              order=cstat["Region"].value_counts().index,
              palette='Blues_r')
sns.set(rc={"figure.figsize": (10, 12)})
```
